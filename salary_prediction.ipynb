# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Reading the file

```python
df = pd.read_csv("survey_results_public.csv");
df.head()
```

```python
df = df[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]
#Renaming column ConvertedCompYearlt to Salary
df = df.rename({"ConvertedCompYearly": "Salary"}, axis=1)
df.head()
```

```python
df = df[df["Salary"].notnull()]
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Removing null data 
df = df.dropna()
df.isnull().sum()
```

```python
#keeping data where person was employed full time
df = df[df["Employment"] == "Employed full-time"]
df = df.drop("Employment", axis=1)
df.info()
```

```python
#clean country data
df['Country'].value_counts()
```

```python
# filtering data from countries with low no of surveys

def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i]>=cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
#salary vs country
fix, ax = plt.subplots(1,1,figsize=(12,7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df["Salary"] <= 200000]
df = df[df["Salary"] >= 10000]
df = df[df['Country'] != 'Other'] #dropping other categories
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df["YearsCodePro"].unique()
```

```python
def clean_experience(x):
    if x == 'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)

df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
df["EdLevel"].unique()
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x or 'Other doctoral' in x:
        return 'Post grad'
    return 'Less than a Bachelors'

df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df["EdLevel"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
df['EdLevel'] = le_education.fit_transform(df['EdLevel'])
df["EdLevel"].unique()
#le.classes_
```

```python
le_country = LabelEncoder()
df['Country'] = le_country.fit_transform(df['Country'])
df["Country"].unique()
```

```python
#split the data
X = df.drop("Salary", axis=1)
y = df["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X, y.values);
```

```python
y_pred = linear_reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y, y_pred))
```

```python
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X, y.values)
```

```python
y_pred = dec_tree_reg.predict(X)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X, y.values)
```

```python
y_pred = random_forest_reg.predict(X);
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X, y.values)
```

```python
regressor = gs.best_estimator_

regressor.fit(X, y.values)
y_pred = regressor.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
# Load and inspect the dataset
df = pd.read_csv("C:/Users/kasoj/Downloads/survey_results_public.csv")
print(df.columns.tolist())
```

```python
# Select and clean relevant columns
df = df[["Country", "ConvertedCompYearly"]].dropna()
df = df.rename(columns={"ConvertedCompYearly": "Salary"})
df = df[(df["Salary"] > 10000) & (df["Salary"] < 250000)]
```

```python
# Boxplot for all countries
fig, ax = plt.subplots(figsize=(12, 7))
df.boxplot(column="Salary", by="Country", ax=ax)
plt.suptitle("Salary (US$) vs Country")
plt.title("")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Boxplot for top 10 countries by response count
top_countries = df["Country"].value_counts().head(10).index
df_top = df[df["Country"].isin(top_countries)]

fig, ax = plt.subplots(figsize=(12, 7))
df_top.boxplot(column="Salary", by="Country", ax=ax)
plt.suptitle("Salary (US$) vs Top 10 Countries")
plt.title("")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
